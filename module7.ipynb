```python
from datetime import datetime

class Pessoa:
    def __init__(self, nome, idade, altura, peso, cor, nacionalidade):
        self.nome = nome
        self.idade = idade
        self.altura = altura
        self.peso = peso
        self.cor = cor
        self.nacionalidade = nacionalidade

    def data_nascimento(self):
        return datetime.now().year - self.idade

class Student(Pessoa):
    def __init__(self, nome, idade, altura, peso, cor, nacionalidade, idStudent):
        Pessoa.__init__(self, nome, idade, altura, peso, cor, nacionalidade)
        self.idStudent = idStudent

class Teacher(Pessoa):
    def __init__(self, nome, idade, altura, peso, cor, nacionalidade, idTeacher):
        Pessoa.__init__(self,nome, idade, altura, peso, cor, nacionalidade)
        self.idTeacher = idTeacher

pessoa = Student("Leonel Caetano", 23, 1.84, 84, "Preta", "Angolana", 1)

listStudent = list()

listStudent.append(pessoa)

pessoa = Student("Vânia Caetano", 15, 1.50, 50, "Preta", "Angolana", 2)

listStudent.append(pessoa)

for x in listStudent:
    print(x.nome ,x.data_nascimento(), x.nacionalidade , x.idStudent, sep=" - ")
```

```python
class Veiculo:
    def __init__(self, peso, potencia, nRodas):
        self.__peso = peso
        self.__potencia = potencia
        self.__nRodas = nRodas

    @property
    def nRodas(self):
        return self.__nRodas

    @property
    def peso(self):
        return self.__peso
    @peso.setter
    def peso(self, value):
        self.__peso = value

    def getPotencia(self):
        return self.__potencia
    def setPotencia(self, value):
        self.__potencia = value

    def trajectoria(self):
        return (self.__peso / self.__potencia) * 1000

    potencia = property(getPotencia, setPotencia)

class Onibus(Veiculo):
    def __init__(self, peso, potencia, nRodas):
        Veiculo.__init__(self, peso, potencia, nRodas)

class Carro(Veiculo):
    def __init__(self, peso, potencia, nRodas):
        Veiculo.__init__(self, peso, potencia, nRodas)

class Moto(Veiculo):
    def __init__(self, peso, potencia, nRodas):
        Veiculo.__init__(self, peso, potencia, nRodas)


carro = Carro(2000, 1500, 4)
onibus = Onibus(5000, 3000, 8)
moto = Moto(200, 300, 2)

carro.potencia = 3000
carro.peso = 2500

print(carro.potencia, carro.nRodas, carro.peso, carro.trajectoria())
print(onibus.peso, onibus.potencia, onibus.nRodas, onibus.trajectoria())
print(moto.peso, moto.potencia, moto.nRodas, moto.trajectoria())
    
```
